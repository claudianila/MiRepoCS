# Scatter plot
La principal diferencia entre plt.scatter y plt.plot es que se puede usar para crear gráficos de dispersión donde
las propiedades de cada punto individual (tamaño, color de cara, color de borde, etc) pueden controlar
o mapear de forma individual a los datos.

```python
plt.scatter(x, y, marker='o');
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,  #alpha es el nivel de transparencia y permite visualizar resultados que se traslapan
cmap='viridis')
plt.colorbar(); # show color scale
```

```python
# Ejemplo con los datos de Iris
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
plt.scatter(features[0], features[1], alpha=0.2,
s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

Esta configuración nos permite explorar 4 cuaracterísticas de los datos, las coordenadas x, y corresponden
la longitud y ancho del sépalo, el tamaño al ancho del pétalo y el color a la especie correspondiente.
Los gráficos de dispersión con multicolor y multiples características como este pueden ser muy útiles para
explorar o presentar los datos

## Visualización de errores
### Gráficos de error básicos

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)
plt.errorbar(x, y, yerr=dy, fmt='.k');
```

Ahora, coloquemos los errores en un color más claro

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
ecolor='lightgray', elinewidth=3, capsize=0);
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
# define the model and draw some data
model = lambda x: x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)
# Compute the Gaussian process fit
#kernel = C(1.0, (1e-3, 1e3)) * RBF(10, (1e-2, 1e2))
gp = GaussianProcessRegressor(kernel=None, n_restarts_optimizer=10) #OJO, esto se tuvo que editar, porque del libro, no corre porque cambió la función
gp.fit(xdata[:,np.newaxis],ydata)
xfit = np.linspace(0,10,1000)
yfit, sigma = gp.predict(xfit[:,np.newaxis], return_std=True)
dyfit = 2*sigma # 2*sigma ~ 95% confidence region
```

```python
# Visualize the result
plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-', color='gray')
plt.fill_between(xfit, yfit - dyfit, yfit + dyfit,
color='gray', alpha=0.2)
plt.xlim(0, 10);
```

```python
df_gptest = df[['Muerto','HOSPITALIZADO','dias_sintomas_ingreso']]
grouped_test1 = df_gptest.groupby(['Muerto','HOSPITALIZADO'],as_index=False).mean()
grouped_test1
```

```python
# Cambiamos las etiquetas
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

#label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```
